```python
# using numpy
import numpy as np

n1 = np.array([1, 2, 3])
n2 = np.array([[1, 2, 3],
               [4, 5, 6]])
n3 = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(n1.shape, n2.shape, n3.shape)
```

```python
#using pytorch
import torch

t1 = torch.tensor([1, 2, 3])

t2 = torch.tensor([[1, 2, 3],
                   [4, 5, 6]])

t3 = torch.tensor([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(t1.shape)
print(t2.shape)
print(t3.shape)
```

```python
# basic operations using numpy
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Addition ",a + b)
print("Subtraction ",a - b)
print("Multiplication ",a * b)
print("Division ",a / b)

print("Dot Product ",np.dot(a, b))
```

```python
# basic operations using pytorch
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

print("Addition ",a + b)
print("Subtraction ",a - b)
print("Multiplication ",a * b)
print("Division ",a / b)

print("Dot Product",torch.dot(a, b))
```

```python
#matrix multiplication using pytorch
A = torch.tensor([[1, 2],
                  [3, 4]])
B = torch.tensor([[5, 6],
                  [7, 8]])

print(torch.matmul(A, B))
```

```python
#matrix multiplication using numpy
A @ B
```

```python
#slicing
print("using pytorch")
t = torch.tensor([[10, 20, 30],
                  [40, 50, 60]])

print(t[0])
print(t[1, 2])
print(t[:, 1])
print("\n")
print("using numpy")
t = np.array([[10, 20, 30],
              [40, 50, 60]])

print(t[0])
print(t[1, 2])
print(t[:, 1])
```

```python
#boolean masking
print("using pytorch")
t = torch.tensor([1, 5, 8, 2, 9])
mask = t > 5
print(mask)
print(t[mask])

print("\nusing numpy")
t = np.array([1, 5, 8, 2, 9])
t[t > 5]
```

```python
#extracting subtensors
t = torch.arange(16).reshape(4, 4)

sub = t[1:3, 1:3]
print(sub)
```

```python
# view in tensors
x = torch.arange(6)
y = x.view(2, 3)
print(y)
```

```python
# reshape in numpy
x.reshape(3, 2)

# reshape in torch
z = x.reshape(3, 2)
```

```python
# squeeze in torch
x = torch.tensor([[1, 2, 3]])

print(x.shape)
print(x.squeeze().shape)
```

```python
# unsqueeze in torch
x = torch.tensor([1, 2, 3])

print(x.shape)
print(x.unsqueeze(0))
print(x.unsqueeze(1))
```

```python
#broadcasting failures
a = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

b = torch.tensor([10, 20, 30])

print(a + b)
```

```python
# inplace operations
x = torch.tensor([1, 2, 3])
y = x + 1

print(x)
```

```python
# outplace operations
x = torch.tensor([1, 2, 3])
x += 1

print(x)
```
